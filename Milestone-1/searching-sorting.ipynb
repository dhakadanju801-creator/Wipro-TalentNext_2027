```python
# Linear Search

# brute force

def linear_search(arr,item):
    for i in range(len(arr)):
        if arr[i] == item:
            return i 
    return -1
```

```python
# Binary Search
# Sorted Array

def binary_search(arr,low,high,item):
    
    print(" low = ",low ,"high = ",high,end=' ' )
    
    if low <= high:
        #search
        mid = (low + high) // 2

        print("mid value is",arr[mid])
        
        if arr[mid] == item:
           return mid
        elif arr[mid] > item:
           return binary_search(arr,low,mid-1,item)
        else:
           return binary_search(arr,mid+1,high,item)
    else:
       return -1
```

```python
arr = [12,24,35,46,57,68,80,99,100]
print(binary_search(arr,0,len(arr)-1,200))
```

```python
# Sorting
def is_sorted(arr):

    sorted = True
    for i in range(len(arr) - 1):
        if arr[i]>arr[i+1]:
            sorted = False
    return sorted

```

```python
arr = [1,2,3,4,5,8,6]
is_sorted(arr)
```

```python
#  BUBBLE SORT                 # space complxity = 1 (constant)
def bubble_sort(arr):
    for i in range(len(arr)-1):
        flag = 0
        for j in range(len(arr)-1 -i):
            if arr[j] > arr[j+1]:
              arr[j],arr[j+1] = arr[j+1],arr[j]
              flag = 1
    
        if flag == 0:
           break

    
    #print(arr)
    return arr
```

```python
a = 5
b = 3
a,b = b,a
print(a)
print(b)
```

```python
arr = [23,12,34,11,100,56,78]
bubble_sort(arr)
```

```python
# SELECTION SORT                                #Time complexity = o(n^2)  , space complexity = o(1)
def selection_sort(arr):

    for i in range(len(arr) -1 ):

       # print(i+1,"pass",end=" ")
        
        min = i

        #print("current min is" ,arr[min])
        
        for j in range(i+1,len(arr)):
            #print("current item under obervation",arr[j])
            
            if arr[j] < arr[min]:
                #print("current item is less than  min")
                min = j
                #print("Now the min has become",arr[min])

        arr[i],arr[min] = arr[min],arr[i]
        #print(arr)
        #print("*"*50)
   # print(arr)
    return arr
```

```python
arr = [23,12,34,11,100,56,78]
selection_sort(arr)
```
