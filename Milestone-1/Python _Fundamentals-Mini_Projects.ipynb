### # Project 1: Travel Suggestion


```python
distance = float(input("How far would you like to travel in miles? "))

if (distance < 3):
    print("Suggest Bicycle to your destination.")
elif (distance < 300):
    print("Suggest Motor-Cycle to your destination.")
else:
    print("Suggest Super-Car to your destination.")
```

### # Project 2: Server Cost Calculator


```python
# Cost of one server per hour
cost_per_hour = 0.51

# Calculate costs
cost_per_day = cost_per_hour * 24
cost_per_week = cost_per_day * 7
cost_per_month = cost_per_day * 30   # Assuming 30 days in a month

# Budget
budget = 918
days_can_run = budget / cost_per_day

# Display results
print("Cost to operate one server per day: $", round(cost_per_day, 2))
print("Cost to operate one server per week: $", round(cost_per_week, 2))
print("Cost to operate one server per month: $", round(cost_per_month, 2))
print("Number of days one server can operate with $918:", round(days_can_run, 2))
```
