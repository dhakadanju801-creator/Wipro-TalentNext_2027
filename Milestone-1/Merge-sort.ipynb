```python
# Merge sort                 # worst case time  complexity = nlogn 
def merge_sorted(arr1,arr2,arr):

    i = j = k= 0
    
    #merged = []
    
    while i < len(arr1) and j < len(arr2):

        if arr1[i] < arr2[j]:
            #merged.append(arr1[i])
            arr[k] = arr1[i]
            i+=1
        else:
            #merged.append(arr2[j])
            arr[k] = arr2[j]
            j+=1
        k+=1
             
    while i < len(arr1):
        #merged.append(arr1[i])
        arr[k] = arr1[i]
        i+=1
        k+=1

    while j < len(arr2):
        #merged.append(arr2[j])
        arr[k] = arr2[j]
        j+=1
        k+=1
            
    #return merged
    return
            
        
 
```

```python
arr1 = [1,2,4]
arr2 = [3,5,7,8,9,10]

merge_sorted(arr1,arr2)
```

```python
def merge_sort(arr):
    
    if len(arr) == 1:
        return arr 
        
    mid = len(arr)//2
    
    left = arr[:mid]
    right = arr[mid:]
    
    #left = merge_sort(left)
    #right = merge_sort(right)
    merge_sort(left)
    merge_sort(right)
    #return merge_sorted(left,right)
    merge_sorted(left,right,arr)


    
```

```python
arr = [2,1,5,8,9,6,7,4]
merge_sort(arr)
print(arr)
```
