```python
# Function to sort hyphen-separated colors
def sort_colors(color_string):
    colors = color_string.split("-")   # Split the string into a list
    colors.sort()                      # Sort alphabetically
    return "-".join(colors)            # Join back with hyphens

# Input from user
color_input = input("Enter hyphen-separated colors: ")

# Display the sorted colors
print("Sorted colors:", sort_colors(color_input))
```
