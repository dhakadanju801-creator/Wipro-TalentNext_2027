```python
#problem1 5
def intTostr(i):
    digits = '0123456789'
    if i == 0:
        return '0'
    result = ''
    while i > 0:                                  # time complexity= o(logn)
        result = digits[i % 10] + result
        i = i // 10
    return result
```

```python
#problem1 11
L=[1,2,3,4,5]
for i in range(0,len(L)//2):
    other = len(L)-i-1
    temp = L[i]
    L[i] = L[other]
    L[other] = temp                    # o(n)
print(L)
```

```python
#problem1 12
def factorial(n):
    if n == 1:
        return 1                        #o(n)
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
#problem1 13
def fib(n):
    if n == 1 or n == 0:
        return 1                                        #o(2^n)
    else:
        return fib(n-1) + fib(n-2)
fib(10)
```

```python
#problem1 14
def power(num):
    if num < 1:
        return 0
    elif num == 1:
        print(1)
        return 1
    else:
        prev = power(num//2)               #o(logn)
        curr = prev*2
        print(curr)         
        return curr
power(45)
        
```

```python
# problem 15
def mod(a,b):
    if b<= 0:
        return -1                #constant time operatin boz no loop and no recursion
    div = a//b
    return a-div*b
mod(5,3)
```

```python
#problem1 16
def sum_digits(num):
    sum = 0
    while(num > 0):
        sum+=num%10
        num/=10                    # o(logn)
    return sum
sum_digits(123)
```
